Peça um número e informe se ele é maior que 10. Crie um algoritmo em Python para fazer essa verificação.

```python
n = float(input('digite um número e eu irei dizer se é maior ou menor que 10:'))
if n > 10:
  print(f'seu número {n} é maior do que 10')
else:
  print(f'seu número {n} é menor do que 10')
```

Peça um número inteiro e informe se ele é par. Crie um algoritmo em Python que verifique se o número é divisível por 2.

```python
n = float(input('digite um número e eu direi se é par ou impar:'))
r = n % 2
if r == 0:
  print ('o número é par.')
else:
  print('o número é impar')
```

Peça um nome e informe se ele contém a letra "a". Crie um algoritmo em Python que faça essa verificação.

```python
n = input('digite um número e eu direi se tem a letra a:')
if "a" in n:
  print(f'tem a em {n}.')
else:
  print(f'não tem a em {n}.')
```

Peça um número e informe se ele é múltiplo de 5. Crie um algoritmo em Python que verifique se o número é divisível por 5.

```python
n = float(input('digite um número e eu direi se ele é multiplo de 5:'))
r = n / 5
if r >= 1:
  print(f'{n} é multiplo de 5.')
else:
  print(f'{n} não é multiplo de 5.')
```

Peça ao usuário para digitar "S" para continuar ou "N" para sair e informe a ação correspondente. Crie um algoritmo em Python que faça essa validação e exiba a mensagem correta.

```python
r = input('digite S para continuar ou N para sair:')
if r == 'n':
  print('você saiu.')
elif r == 's':
  print('você continuou.')
else:
  print('não entendi.')
```

Peça um número e informe se ele é negativo, zero ou positivo. Crie um algoritmo em Python para classificar o número.

```python
n = float(input('digite um número e eu direi se é negativo ou positivo:'))
if n > 0:
  print(f'{n} é positivo.')
elif n == 0:
  print('este é o zero.')
else:
  print(f'{n} é negativo')
```

Peça um número de 1 a 7 e informe o dia da semana correspondente. Crie um algoritmo em Python que associe o número ao dia correto.

```python
n = input('digite um núnero e eu direi o dia da semana correspondente.')
if n == "1":
  print('domingo')
elif n == '2':
  print('segunda')
elif n == "3":
  print('terça')
elif n == '4':
  print('quarta')
elif n == "5":
  print('quinta')
elif n == '6':
  print('sexta')
elif n == '7':
  print('sabado')
else:
  print('esse não é um dia da semana.')
```
